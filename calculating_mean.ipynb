```python
import numpy as np
import pandas as pd
```

```python
raw_data = pd.read_csv(r"C:\Users\amdm_\Documents\GitHub\viscosity_liquid_transfer_Pablo\Sartorious_experiments\Experimental_Data\Best_parameters_data\Batch_optimization\Viscosity_std_817_BP_exp3.csv", header=0)
raw_data
```

```python
raw_data["act_volume"] = (raw_data["m"] / raw_data["density"] * 1000)
raw_data["abs_error"] = (raw_data["volume"] - raw_data["act_volume"])
raw_data["act_time"] = ((raw_data["volume"] / raw_data["aspiration_rate"]) + (raw_data["volume"] / raw_data["dispense_rate"]) + raw_data["delay_aspirate"] + raw_data["delay_dispense"])
raw_data.head()
```

```python
grouped = raw_data.groupby(by=["volume"])
```

```python
data_1000 = pd.DataFrame(grouped.get_group(1000))
mean_1000 = pd.DataFrame(data_1000.mean())
mean_1000 = mean_1000.transpose()
```

```python
mean_1000
```

```python
final_1000 = mean_1000[["volume", "act_volume", "abs_error", "act_time"]]
final_1000["std"] = data_1000["act_volume"].describe().loc["std"]
final_1000
```
