```python
df_experiment["act_volume"] = (df_experiment["m"] / df_experiment["density"] * 1000)
df_experiment["abs_error"] = (df_experiment["volume"] - df_experiment["act_volume"])
df_experiment["act_time"] = ((df_experiment["volume"] / df_experiment["aspiration_rate"]) + (df_experiment["volume"] / df_experiment["dispense_rate"]) + df_experiment["delay_aspirate"] + df_experiment["delay_dispense"])
df_experiment["percent_error"] = (df_experiment["abs_error"] / 10)
df_experiment.head()
```

```python
df_summary_all = pd.DataFrame()
for x in df_experiment['volume'].unique():
    df_experiment_1000 = df_experiment.where(df_experiment['volume'] == x).dropna(how='all')
    df_summary = pd.DataFrame(columns = (f'{x} Actual Volume', f'{x} Absolute Error', f'{x} Percentage Error', f'{x} Actual Time', f'{x} Std Dev', f'{x} Relative Std Dev') )
    data = [df_experiment_1000['act_volume'].mean(), df_experiment_1000['abs_error'].mean(), df_experiment_1000['percent_error'].mean(), df_experiment_1000['act_time'].mean(), df_experiment_1000['act_volume'].std(), (df_experiment_1000['act_volume'].std() / df_experiment_1000['act_volume'].mean() * 100)]
    df_summary.loc[df_experiment['liquid'].iloc[0]] = data
    df_summary_all = pd.concat([df_summary_all, df_summary], axis = 1)
```
